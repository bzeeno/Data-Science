```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
data_path = "data/"
lowest_data = pd.read_csv(data_path+"lowest_quintile.csv")
second_data = pd.read_csv(data_path+"second_quintile.csv")
middle_data = pd.read_csv(data_path+"middle_quintile.csv")
fourth_data = pd.read_csv(data_path+"fourth_quintile.csv")
highest_data = pd.read_csv(data_path+"highest_quintile.csv")
```

```python
highest_data.head()
```

## Before Taxes

```python
plt.figure(figsize=(10,10))

plt.plot(highest_data["year"], highest_data["inc_before_transfers_taxes"], label="Highest Quintile")
plt.plot(fourth_data["year"], fourth_data["inc_before_transfers_taxes"], label="Fourth Quintile")
plt.plot(middle_data["year"], middle_data["inc_before_transfers_taxes"], label="Middle Quintile")
plt.plot(second_data["year"], second_data["inc_before_transfers_taxes"], label="Second Quintile")
plt.plot(lowest_data["year"], lowest_data["inc_before_transfers_taxes"], label="Lowest Quintile")

plt.title("Income By Quintile", size=15)
plt.xlabel("Year", size=15)
plt.ylabel("Income Before Taxes [$]", size=15)
plt.legend()
```

```python
plt.figure(figsize=(10,10))

plt.plot(highest_data["year"], highest_data["inc_after_transfers_taxes"], label="Highest Quintile")
plt.plot(fourth_data["year"], fourth_data["inc_after_transfers_taxes"], label="Fourth Quintile")
plt.plot(middle_data["year"], middle_data["inc_after_transfers_taxes"], label="Middle Quintile")
plt.plot(second_data["year"], second_data["inc_after_transfers_taxes"], label="Second Quintile")
plt.plot(lowest_data["year"], lowest_data["inc_after_transfers_taxes"], label="Lowest Quintile")

plt.title("Income By Quintile", size=15)
plt.xlabel("Year", size=15)
plt.ylabel("Income After Taxes [$]", size=15)
plt.legend()
```
