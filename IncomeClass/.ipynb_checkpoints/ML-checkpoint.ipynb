### Logistic Regression

```python
# Test if data needs to be transformed
def PlotFeaturesTargets(features_to_plot,targets_to_plot,data_to_plot):
    fig = plt.figure(figsize=(18,30))
    
    num_features = features_to_plot.shape[1] 

    # will have 3 cols, so to calculate number of rows we need:
    num_rows = np.ceil( num_features/3. ) 


    for i in range(num_features):
        axis = fig.add_subplot(num_rows, 3, i+1)
        plot = sns.scatterplot(x=features_to_plot.iloc[:,i], y=targets.iloc[:,0], data=data_to_plot, ax=axis)
    plt.show()
    
PlotFeaturesTargets(features, targets, df)
```

```python
clf = LogisticRegression(random_state=42)
scores = cross_val_score(clf, features, targets, cv=10)
print(scores)
```
