```python
import numpy as np
import pandas as pd

from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import LogisticRegression

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn import svm

from sklearn.model_selection import cross_val_score

from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
# from sklearn.metrics import roc_auc_score
from sklearn.metrics import make_scorer
from sklearn.metrics import classification_report

import tensorflow as tf
from sklearn.model_selection import KFold

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data_path = "Data/"
raw_df = pd.read_csv(data_path+"preprocessed.csv")
raw_df.head()
```

# Separate features and targets

```python
df = raw_df.copy()
features = df.iloc[:,:-1]
targets = df.iloc[:,-1:]
```

```python
features.head()
```

```python
targets.head()
```

# Train Test Split

```python
features_train, features_test, targets_train, targets_test = train_test_split(features, targets, test_size=0.1, random_state=42)
```

```python
# Flatten targets data
targets_train = np.ravel(targets_train)
targets_test = np.ravel(targets_test)
```

# ML

```python
scorer = make_scorer(accuracy_score) 
```

### Logistic Regression

```python
clf_logistic = LogisticRegression(random_state=42)
scores = cross_val_score(clf_logistic, features_train, targets_train, cv=10, scoring=scorer)
print(scores)
```

```python
clf_logistic.fit(features_train,targets_train)
```

```python
pred_logistic = clf_logistic.predict(features_test)
acc = accuracy_score(pred_logistic, targets_test)
print(acc)
```

### Adaboost

```python
# use Area Under ROC Curve (AUROC) OR balanced_accuracy_score
# Set hyperparameters to test
hyper_parameters_AB = {"n_estimators": [25,50,100],
                       "base_estimator": [DecisionTreeClassifier(max_depth=2),DecisionTreeClassifier(max_depth=3),RandomForestClassifier(max_depth=2)]
                      }
# Create GridSearchCV object
grid_AB = GridSearchCV(AdaBoostClassifier(), param_grid=hyper_parameters_AB, scoring=scorer)
fit_grid_AB = grid_AB.fit(features_train,targets_train) # fit grid to our training data
best_score_AB = fit_grid_AB.best_score_
clf_AB = fit_grid_AB.best_estimator_ # save the best performing classifier to variable

print("Best Hyper-parameters: \n", grid_AB.best_params_)
print("\n--------------------------------------------------------------------------------------------------\n")
print("Model after hyper-parameter tuning: \n",grid_AB.best_estimator_)
print("\n--------------------------------------------------------------------------------------------------\n")
print("Score: ", best_score_AB)
```

```python
# Evaluation
pred_AB = clf_AB.predict(features_test)
eval_score_AB = accuracy_score(pred_AB, targets_test)
print(eval_score_AB)
```

### SVC

```python
# SVC is a penalized learning algorithm,  so it will increase the cost of missclassifications on the minority class
# This will help mitigate the unbalanced data
#Create a svm Classifier
hyper_parameters_SVC = [
                        {"C": np.logspace(start=-3, stop=2, num=6), "kernel": ["linear"]},
                        {"C": np.logspace(start=-3, stop=2, num=6), "kernel": ["rbf"], "gamma": np.logspace(start=-3, stop=2, num=6)}
                       ]

# Create GridSearchCV object
grid_SVC = GridSearchCV(svm.SVC(), param_grid=hyper_parameters_SVC, scoring=scorer)

fit_grid_SVC = grid_SVC.fit(features_train,targets_train) # fit grid to our training data

best_score_SVC = fit_grid_AB.best_score_ # Score of best estimator
clf_SVC = fit_grid_SVC.best_estimator_   # save the best performing classifier to variable

print("Best Hyper-parameters: \n", grid_SVC.best_params_)
print("\n--------------------------------------------------------------------------------------------------\n")
print("Model after hyper-parameter tuning: \n",clf_SVC)
print("\n--------------------------------------------------------------------------------------------------\n")
print("Score: ", best_score_SVC)
```

```python
pred_SVC = clf_SVC.predict(features_test)
eval_score_SVC = accuracy_score(pred_SVC, targets_test)
print(eval_score_SVC)
```

# Deep Learning

```python
# First, we need the data to be saved in npz format
np.savez(data_path+"train_validation_data", inputs=features_train, targets=targets_train)
np.savez(data_path+"test_data", inputs=features_test, targets=targets_test)
```

```python
npz = np.load(data_path+"train_validation_data.npz")
train_validation_inputs = npz["inputs"]
train_validation_targets = npz["targets"]

npz = np.load(data_path+"test_data.npz")
test_inputs = npz["inputs"]
test_targets = npz["targets"]
```

```python
print("Shape of train/validation inputs:",np.shape(train_validation_inputs))
print("Shape of train/validation targets:",np.shape(train_validation_targets))
print("\n")
print("Shape of test inputs:",np.shape(test_inputs))
print("Shape of test targets:",np.shape(test_targets))
```

```python
# Neural Network shape
INPUT_SIZE = 103
HIDDEN_LAYER_SIZE = 100
OUTPUT_SIZE = 1
```

```python
# Num of folds in K-fold CV
num_folds = 10
# lists to hold accuracy and loss of each fold
acc_fold = []
loss_fold = []

# Fraction of combined train/validation data for validation
validation_split = 0.1

kf = KFold(n_splits=num_folds, shuffle=True, random_state=42)

batch_size = 100
num_epochs = 300

fold_num = 1 # which fold the for loop is on
for train, validate in kf.split(train_validation_inputs, train_validation_targets):
    network = tf.keras.models.Sequential() # Begin Neural Network
    network.add(tf.keras.layers.Flatten())
    network.add(tf.keras.layers.Dense(INPUT_SIZE, activation="sigmoid"))
    network.add(tf.keras.layers.Dense(HIDDEN_LAYER_SIZE, activation="tanh"))
    network.add(tf.keras.layers.Dense(HIDDEN_LAYER_SIZE, activation="relu"))
    network.add(tf.keras.layers.Dense(OUTPUT_SIZE, activation="sigmoid"))
    network.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
    
    # Generate a print
    print('------------------------------------------------------------------------')
    print(f'Training for fold {fold_num} ...')
    
    # Fit data to model
    fit_model = network.fit(train_validation_inputs[train], train_validation_targets[train],
                            batch_size=batch_size,
                            epochs=num_epochs,
                            verbose=2,
                            validation_split=validation_split)
    
    # Metrics. evaluate method returns: [loss, accuracy]
    validation_scores = network.evaluate(train_validation_inputs[validate], train_validation_targets[validate], verbose=0)
#     print(f'Score for fold {fold_num}: {model.metrics_names[0]} of {validation_scores[0]}; {model.metrics_names[1]} of {validation_scores[1]*100}%')
    acc_fold.append(validation_scores[1])
    loss_fold.append(validation_scores[0])

    # Increase fold number
    fold_num = fold_num + 1
    
    
# == Provide average scores ==
print('------------------------------------------------------------------------')
print('Score per fold')

for i in range(0, len(acc_fold)):
    print('------------------------------------------------------------------------')
    print(f'> Fold {i+1} - Validation Loss: {loss_fold[i]} - Validation Accuracy: {acc_fold[i]}')
    
print('------------------------------------------------------------------------')
print('Average scores for all folds:')
print(f'> Accuracy: {np.mean(acc_fold)} (+- {np.std(acc_fold)})')
print(f'> Loss: {np.mean(loss_fold)}')
print('------------------------------------------------------------------------')
```

```python
test_loss, test_accuracy = network.evaluate(test_inputs, test_targets)
print("Loss: ", test_loss, "\nAccuracy: ", test_accuracy)
```
