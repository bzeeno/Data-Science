```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import PlotFunctions
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

sns.set()
```

```python
# List of column names
column_names = ["age","work class","fnlwgt","education","education-num","marital-status","occupation",
                "relationship","race","sex","capital-gain","capital-loss","hours-per-week",
                "native-country", "income class"]
```

Merge the test and train datasets since I'll probably use cross validation rather than a train-test split

```python
data_path = "Data/"
raw_data1 = pd.read_csv(data_path+"adult.data", names=column_names,skipinitialspace=True)
raw_data2 = pd.read_csv(data_path+"adult.test", names=column_names,skipinitialspace=True)
raw_data_combined = pd.concat([raw_data1, raw_data2])
raw_data_combined.reset_index(drop=True, inplace=True)
raw_data_combined.head()
```

```python
raw_data_combined.shape
```

```python
data = raw_data_combined.copy()
```

```python
data.describe()
```

# Deal with missing data

```python
# Look for how missing values are denoted
for col in data:
    print(np.unique(getattr(data,col).values))
```

```python
indices_to_drop = []

for col in data:
    for i,val in enumerate(data[col]):
        if(val == '?' or val == ' ?'):
            indices_to_drop.append(i)

indices_to_drop = list(set(indices_to_drop))
print(indices_to_drop)
print(len(indices_to_drop))
```

```python
print("Percent to drop:\n", (len(indices_to_drop) / data.shape[0]) * 100, "%")
```

This is a large portion of data to drop, but I think it will be worth it. Using cross-validation will help mitigate this loss of data.

```python
data_no_mv = data.copy()
```

```python
data_no_mv = data_no_mv.drop(indices_to_drop, axis=0)
data_no_mv = data_no_mv.reset_index(drop=True)
data_no_mv
```

```python
data_no_mv.shape
```

# Convert Targets to Categorical

We'll perform one hot encoding on the features after our exploratory data analysis

```python
categorical_df = data_no_mv.copy()
categorical_df.head()
```

```python
# Need to take into account the '.' at the end of some of the targets
categorical_df["income class"] = categorical_df["income class"].map({"<=50K": 0, "<=50K.": 0, ">50K": 1, ">50K.": 1})
print(np.unique(categorical_df["income class"])) # make sure there are only 0's and 1's

categorical_df.head()
```

### Check if data is balanced.

```python
sns.countplot(x=categorical_df["income class"],data=categorical_df)
```

This dataset is unbalanced. I would oversample the 1's, but I will first try other methods in the ML notebook as I would prefer to not change or augment the data if I dont have to. 

```python
# Determine ratio of 1's
# ratio 1:0 = num of 1's / num of 0's
ratio_1s_0s = sum(categorical_df["income class"]) / (categorical_df.shape[0]-sum(categorical_df["income class"]))
print(ratio_1s_0s)
```

### See how each feature affects target

```python
categorical_names = ["work class", "marital-status", "occupation", "relationship", "race", "sex", "native-country"]
```

We're going to separate the dataset into two where each one only has 0's or 1's for income class. This will be temporary; we're going to use these datasets to plot countplots so we can visually compare how each class is divided in the features.

```python
df_majority = categorical_df[categorical_df["income class"]==0] # DF with only income class = 0
df_minority = categorical_df[categorical_df["income class"]==1] # DF with only income class = 1
df_majority = df_majority.sample(frac=(ratio_1s_0s)).reset_index(drop=True) # Only keep ratio of 1:0 so they will be equal

print(df_majority.shape)
print(df_minority.shape)
```

```python
# Temporary DF for countplots
temp_df = pd.concat([df_majority, df_minority], ignore_index=True)
print(temp_df.shape)
temp_df
```
