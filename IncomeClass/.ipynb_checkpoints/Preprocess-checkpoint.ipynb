# Deal with missing data

```python
# Look for how missing values are denoted
for col in data:
    print(np.unique(getattr(data,col).values))
```

```python
indices_to_drop = []

for col in data:
    for i,val in enumerate(data[col]):
        if(val == '?' or val == ' ?'):
            indices_to_drop.append(i)

indices_to_drop = list(set(indices_to_drop))
print(indices_to_drop)
print(len(indices_to_drop))
```

```python
print("Percent to drop:\n", (len(indices_to_drop) / data.shape[0]) * 100, "%")
```

This is a large portion of data to drop, but I think it will be worth it. Using cross-validation will help mitigate this loss of data.

```python
data_no_mv = data.copy()
```

```python
data_no_mv = data_no_mv.drop(indices_to_drop, axis=0)
data_no_mv = data_no_mv.reset_index(drop=True)
data_no_mv
```

```python
data_no_mv.shape
```

### Check if data is balanced.

```python
sns.countplot(x=categorical_df["income class"],data=categorical_df)
```

This dataset is unbalanced. I would up-sample the 1's, but I will first try other methods in the ML notebook as I would prefer to not change or augment the data if I dont have to. 

Update: The models I implemented did not perform as well as I had hoped. I will create a new dataframe that will up-sample the 1's. However, I will do this after the EDA section. 

```python
# Determine ratio of 1's
# ratio 1:0 = num of 1's / num of 0's
ratio_1s_0s = sum(categorical_df["income class"]) / (categorical_df.shape[0]-sum(categorical_df["income class"]))
print(ratio_1s_0s)
```

### See how each feature affects target

```python
categorical_names = ["work class", "marital-status", "occupation", "relationship", "race", "sex", "native-country"]
```

We're going to separate the dataset into two where each one only has 0's or 1's for income class. This will be temporary; we're going to use these datasets to plot countplots so we can visually compare how each class is divided in the features.

```python
df_majority = categorical_df[categorical_df["income class"]==0] # DF with only income class = 0
df_minority = categorical_df[categorical_df["income class"]==1] # DF with only income class = 1
# Only keep ratio of 1:0 so they will be equal
# sample method shuffles rows, which is what we want
df_majority = df_majority.sample(frac=(ratio_1s_0s), random_state=42).reset_index(drop=True) 
print(df_majority.shape)
print(df_minority.shape)
```

```python
# Temporary DF for countplots
temp_df = pd.concat([df_majority, df_minority], ignore_index=True)
print(temp_df.shape)
temp_df
```

```python
# We won't be able to perfectly see the absolute differences between the income classes as we have removed a 
# large portion of the data for these plots. But, we will be able to get an idea of the relative differences
 
PlotFunctions.CategoryTargetCountPlot(temp_df[categorical_names],temp_df["income class"])
```

Some of the biggest takeaways are that most people who make over \\$50k are married. In addition, the most common occupations of those that make over \\$50k are prof-specialty and exec-managerial. Also, the blacks disproportionately make less than \\$50k when compared to whites and asian-pac-islanders. Females also disproportionately make less than \\$50k compared to males

# Balance Dataset

Now, let's up-sample the 1's so that the data will be balanced

```python
df_majority = categorical_df[categorical_df["income class"]==0] # DF with only income class = 0
df_minority = categorical_df[categorical_df["income class"]==1] # DF with only income class = 1

# Upsample minority class
df_minority_upsampled = resample(df_minority, 
                                 replace=True,     # sample with replacement
                                 n_samples=df_majority.shape[0],    
                                 random_state=42)  

# Concatenate majority df and upsampled minority df
upsampled_df = pd.concat([df_majority, df_minority_upsampled])
```

```python
# Show new class counts
print("Minority Class Count: ", sum(upsampled_df["income class"]))
print("Majority Class Count:" , upsampled_df.shape[0]-sum(upsampled_df["income class"]))
```

We'll continue the remainder of the project using the upsampled df.

# One-hot Encoding for Categorical Features

```python
encoded_df = pd.get_dummies(upsampled_df)
encoded_df.head()
```

```python
encoded_df.shape
```

# Standardize Numerical Features

When we standardize the features we will lose the meaning of the numerical values, but it's not too big of a lose since what we really care about is what the most important features are for predicting their income class.

```python
features_to_scale = ["age","fnlwgt","capital-gain", "capital-loss","hours-per-week"]

scaled_features = StandardScaler().fit_transform(encoded_df[features_to_scale]) # Scale features

# make scaled_features into pandas dataframe
scaled_features_df = pd.DataFrame(scaled_features,index=encoded_df.index, columns=features_to_scale) 

# Concatentate encoded_df and scaled_features
scaled_features_df = pd.concat([scaled_features_df, encoded_df.drop(features_to_scale, axis=1)], axis=1)
scaled_features_df
```
